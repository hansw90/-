# request & beautiful soup


### request : HTTP for Humans
- 간결한 코드로 다양한 HTTP 요청을 할 수 있는 라이브러리
- Javascript 처리가 필요한 경우 selenium 을 사용 가능, 하지만 request 에서도 적용이 가능한 부분이 있으며, 이는 request 이용시 크롤링 할 페이지에 대해 다방면의 검토가 필요하다고 볼 수 있다.
- 크롤링을 할때 요청에서 requests를 사용한다면 가장 효율적인 처리라고 볼 수 있다

```python
import requests
 
response = requests.get('http://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cnt&date=20170714')
 
html = response.text
print(html)
```

```python
# 위의 결과물에서 BeautifulSoup를 통해 원하는 랭킹만 뽑아보자
# div class="tit3">
```

```python
import requests

response = requests.get('http://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cnt&date=20170714')

html = response.text

from bs4 import BeautifulSoup
soup = BeautifulSoup(html,'html.parser') #html.parser 를 사용해서 soup에 넣는다
soup.select('div[class=tit3]')
```

```python
for tag in soup.select('div[class=tit3]'):
    print(tag.text.strip()) #출력할떄 공백을 제거 하기 위해 .strip() 을 사용한다
```
